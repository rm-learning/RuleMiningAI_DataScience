### Questão 6

O artigo aborda os principais aspectos relacionados à mineração por regras de associação. Inicialmente, há um levantamento do conceito e relevância da mineração por regras de associação na descoberta de relações entre os dados. O destaque é a apresentação dos três modos de disponibilidade: balanceado, criativo e preciso, que são as diferentes abordagens para lidar com pesquisas científicas. Além disso, a seção fornece informações valiosas sobre as ferramentas pré-definidas para pesquisa na web e arte gráfica, que são recursos essenciais nesse contexto.

Há também uma preocupação em explicar as medidas básicas de avaliação da qualidade das regras de associação, chamadas suporte e confiança. Os dados são apresentados em tabelas simples para visualizar as métricas essenciais. Com isso é possível identificar uma classificação não hierárquica dos algoritmos em três categorias principais: baseados em conjuntos frequentes, baseados em padrões sequenciais e baseados em padrões estruturados.

A comparação dos algoritmos em termos de desempenho, complexidade, aplicações e limitações é apresentada de forma conjunta com tabelas e gráficos e ilustra os resultados de forma visual, tornando mais fácil a compreensão das diferenças e semelhanças entre os algoritmos. A abordagem baseada em conjuntos frequentes inclui o algoritmo a priori, e procura por conjuntos de itens que ocorrem com uma frequência maior ou igual a um limiar mínimo definido pelo usuário. Enquanto a abordagem baseada em padrões sequenciais procura por padrões onde a ordem dos itens é importante. E, por fim, a abordagem baseada em padrões estruturados procura por padrões que envolvem relações entre diferentes tipos de itens ou entre itens e suas propriedades.
